```python
# Set up the matplotlib figure
fig, axes = plt.subplots(2, 2, figsize=(12, 7))

a = tips_df["tip"].values
b = tips_df["size"].values
c = tips_df["total_bill"].values


#1

sns.distplot(a, color = 'red', ax = axes[0,0])
axes[0,0].set_title = 'tip'

#2
sns.distplot(b, color = 'red', ax=axes[0,1])

#3
sns.distplot(c, color = 'red', ax=axes[1,0])

#4
sns.boxplot(x = "tip", y = "day", data = tips_df, color = "#ffcc99", ax=axes[1,1])
sns.swarmplot(x = "tip", y="day", data = tips_df, color = "#ff9900", ax=axes[1,1])
```

HEAT MAP

```python
pc = tips_df[["total_bill","size","tip", "sex"]].corr(method ="pearson")

cols = ["total_bill CC","size CC","tip"]

ax = sns.heatmap(pc, annot=True,
                 yticklabels=cols,
                 xticklabels=cols,
                 annot_kws={'size': 30})
```

```python
ax = sns.heatmap(pc, annot=True,
                 yticklabels=cols,
                 xticklabels=cols,
                 annot_kws={'size': 30},
                cmap="BuPu")
```
